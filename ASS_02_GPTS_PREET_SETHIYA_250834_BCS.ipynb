```python
from transformers import pipeline

# 1. Load the model pipeline
# We use 'text-generation' task. 'gpt2' is the model name.
# device=0 tells it to use the GPU we enabled.
generator = pipeline('text-generation', model='gpt2', device=0)

# 2. Your Prompt
my_prompt = "The legal system in india is quite complex and cumbersome ,thats why ai can simplify it"

# 3. Generate text
# max_length limits the output size.
output = generator(my_prompt, max_length=100, num_return_sequences=1)

# 4. Print result
print(output[0]['generated_text'])
```
